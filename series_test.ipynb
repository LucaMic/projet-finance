```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.figure(figsize=(20, 5))
x = np.linspace(0, 10, 1000)

plt.subplot(1, 2, 1)
plt.plot(x, np.random.randint(-10, 10) * (np.sin(x) + 0.1 * np.random.normal(0, 1, 1000)));

plt.subplot(1, 2, 2)
plt.plot(x, np.random.randint(-10, 10) * (np.sin(x) + np.random.normal(0, 1, 1000)));
plt.show()
```

```python
data = pd.read_csv('CAC.txt', sep=';', header=None)
data.columns = ['id', 'date', 'debut', 'max', 'min', 'fin', 'volume']
plt.figure(figsize=(15, 5))
plt.title('CAC 40')
plt.plot(data['date'], data['max'])
plt.ylabel('Valeur maximum')
plt.xlabel('Date')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
x = np.linspace(0, 100, 100)

fig, axes = plt.subplots(5, figsize=(20, 20))
for i in range(0, 5):
    axes[i].plot(x, np.random.normal(0, 1, 100))
    
```

```python
fig, axes = plt.subplots(5, figsize=(20, 20))
for i in range(0, 5):
    tmp = np.random.randint(-10, 10)
    axes[i].plot(x, tmp * x + 100*np.random.normal(0, 1, 100))
    axes[i].plot(x, tmp * x)
```
